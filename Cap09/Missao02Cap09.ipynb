```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

## Número de veículos pertencentes a cada marca

```python
df['brand'].value_counts()
```

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r")
order = df['brand'].value_counts()
g = sns.catplot(y="brand", data=df, kind="count", edgecolor=".4", palette=palette, height=7,
                aspect=1.5)
g.ax.xaxis.set_label_text("Número de Veículos",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Marca",fontdict= {'size':16})
plt.show()
```

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r")
order = df['brand'].value_counts()
g = sns.catplot(y="brand", data=df, kind="count", edgecolor=".4", palette=palette, height=7,
                aspect=1.5)
g.ax.xaxis.set_label_text("Número de Veículos",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Marca",fontdict= {'size':16})
plt.show()
```

```python
df.head(5)
```

```python
sns.set_theme(style="whitegrid")
colors = ["#00e600", "#ff8c1a","#a180cc"]
fig, ax = plt.subplots(figsize=(10,5))

sns.barplot(data=df,x="vehicleType", y="price", hue="gearbox", palette=colors, alpha=.8)
ax.set_title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio",
             fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de veiculo",fontdict= {'size':16})
ax.yaxis.set_label_text("Preco medio",fontdict= {'size':16})
plt.show()
```
