```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
df.head()
```

```python
df.columns
```

```python
sns.set_theme(style="whitegrid")
colors = ["#e683d8","#49be25",'#2596be']
fig, ax = plt.subplots(figsize=(10,5))

sns.barplot(data=df,x="fuelType", y="price", hue="gearbox", palette=colors)
ax.set_title("Preço médio dos veículos por tipo de veículo por tipo de combustivel",
             fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de combustivel",fontdict= {'size':16})
ax.yaxis.set_label_text("Preco medio",fontdict= {'size':16})
plt.show()
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
df.head(3)
```

```python
df.columns
```

```python
sns.set_theme(style="whitegrid")
colors = ["#1142D8","#e8ac13",'#BDBDB9']
fig, ax = plt.subplots(figsize=(10,7))

sns.barplot(data=df,x="vehicleType", y="powerPS", hue="gearbox", palette=colors)
ax.set_title("Preço médio dos veículos por tipo de veículo por cavalos de potencia",
             fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de combustivel",fontdict= {'size':16})
ax.yaxis.set_label_text("Potencia media",fontdict= {'size':16})
plt.show()
```
