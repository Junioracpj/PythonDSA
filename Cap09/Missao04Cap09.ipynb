```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df2 = df.groupby('vehicleType').mean()
df2.head()
```

```python
df["brand"].unique()
```

```python
trial = pd.DataFrame()
brand = []
tipo = []
media = []
for b in list(df["brand"].unique()):
    for v in list(df["vehicleType"].unique()):
        media.append(df[(df["brand"] == b) & (df["vehicleType"] == v)]["price"].mean())
        tipo.append(v)
        brand.append(b)
```

```python
df = pd.DataFrame()
df['brand'] = brand
df['tipo'] = tipo
df['media'] = media
```

```python
df['media'].fillna(0,inplace=True)
```

```python
df["media"] = df["media"].astype(int)
```

```python
df.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
sns.set_theme()

glue = df.pivot("brand", "tipo", 'media')

f, ax = plt.subplots(figsize=(15,20))

sns.heatmap(glue, annot=True, fmt="d", linewidths=.5, ax=ax)

ax.set_title("Heatmap - Preço médio de um veículo por marca e tipo de veículo",fontdict={'size':20})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':20})
ax.yaxis.set_label_text("Marca",fontdict= {'size':20})

plt.show()
```
