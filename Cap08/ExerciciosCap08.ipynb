### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond.
# Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
for i in range(len(cond)):
    if cond[i] == True:
        print(xarr[i])
    else:
        print(yarr[i])
```

### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
import os
import sqlite3

# Criando conexao
con = sqlite3.connect('teste2.db')
# Criando o cursor
cur = con.cursor()
```

```python
sql_create = 'create table Exemplo '\
'(id integer primary key, '\
'nome varchar(100), '\
'categoria varchar(140),'\
'valor intenger(99999))'

cur.execute(sql_create)
```

```python
import numpy as np

recset = [(1000, 'Ciência de Dados', 'Data Science', ''),
          (1001, 'Big Data Fundamentos', 'Big Data','' ),
          (1002, 'Python Fundamentos', 'Análise de Dados','')]
sql_insert = 'insert into Exemplo values (?, ?, ?, ?)'
for rec in recset:
    cur.execute(sql_insert, rec)
con.commit()
con.close()
```

```python
import pandas as pd

conn2 = sqlite3.connect("teste2.db")
tabelas = pd.read_sql_query("select * from Exemplo", conn2)
display(tabelas)
```

```python
tabelas['valor'] = np.random.rand(3)
display(tabelas)
```
